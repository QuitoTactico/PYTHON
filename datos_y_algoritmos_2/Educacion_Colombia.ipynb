3. Generadores de información

```python
def generar_info() -> dict :
    # dependiendo del sexo, le asigna un nombre
    sexo = random.choice(['Masculino','Femenino'])

    if sexo =='Masculino':
        nombre = random.choice(nombreHombre)
    else:
        nombre = random.choice(nombreMujer)

    apellido1   = random.choice(apellidos)
    apellido2   = random.choice(apellidos)
    edad        = random.randint(8, 25)
    departamento_code       = random.choice(departamentos_keys)
    departamento            = departamentos[departamento_code]
    municipio               = random.choice(municipios[departamento_code])
    indice_desercion_dep    = indices_desercion_dep[departamento_code]

    info = {'nombre'                : f'{nombre} {apellido1} {apellido2}', 
            'edad'                  : edad, 
            'sexo'                  : sexo, 
            'departamento_code'     : departamento_code, 
            'departamento'          : departamento,
            'municipio'             : municipio,
            'indice_desercion_dep'  : indice_desercion_dep                }   
    return info



def generar_historial_academico(indice_desercion : float, edad : int) -> dict:
    # -1 = Párvulos , 0 = Jardín , [1-6] = Primaria , [7-11] = Bachillerato
    grado = -1
    repitencias_y_deserciones = 0
    desertando = False

    # Normalmente se comienza a estudiar párvulos a los 5 años, pero podría suceder luego
    # La lista [0]*6 + [1]*3 + [2] emula esas probabilidades de entrar después (hasta 2 años)
    año                     = 2022- edad + 5 + random.choice([0]*6 + [1]*3 + [2])
    probabilidad_repetir    = 5  # En % (1 de cada 20 personas perdió 1er grado).

    # Deserción Temporal = Que no siga cursando por cierto tiempo (puede o no ser hasta la actualidad). 
    # Eso no significa deserción total, que es lo que el algoritmo final intenta predecir.
    # La prob de deserción temporal comienza desde el índice de deserción real de su departamento.
    # Y cada año, la probabilidad de volver a estudiar luego de una deserción es del 50%
    probabilidad_desercion_temporal = indice_desercion 
    probabilidad_regresar_de_desercion = 50

    historial_academico = deque() 

    # 11 no está incluído. La base de datos es de personas aún en proceso educativo.
    while grado <= 10 and  año <= 2022:

        # Ver si se sale del colegio temporalmente. 
        if random.randint(0,100) < probabilidad_desercion_temporal and desertando == False:
            desertando = True
            repitencias_y_deserciones += 1
            # La pérdida de hábitos de estudio hace más probable el perder el año
            probabilidad_repetir += 10  
            historial_academico.append([año,grado,'Desertó temporalmente'])
            año += 1
            continue

        # Ver si vuelve al colegio y cursa el año  |  La prob es de 50% constante
        if random.randint(0,100) < probabilidad_regresar_de_desercion and desertando == True:
            desertando = False
            # Volver a la educación reduce la probabilidad de que vuelva a desertar
            probabilidad_desercion_temporal -= 5

        # Ver si pierde el año
        if random.randint(0,100) < probabilidad_repetir and desertando == False:
            # repitió año, eso aumenta su probabilidad de perder de nuevo o desertar
            repitencias_y_deserciones += 1
            probabilidad_repetir += 10
            probabilidad_desercion_temporal += 3
            historial_academico.append([año,grado,'Reprobó'])
        
        # No perdió el año, pero...
        else:
            # Ganándolo
            if desertando == False:
                historial_academico.append([año,grado,'Aprobó'])
                grado += 1
            # Desertándolo
            if desertando == True:
                historial_academico.append([año,grado,'Desertó temporalmente'])
                repitencias_y_deserciones += 1
        
        año += 1

    # retornaremos esta información para agregarla a la Persona.
    historial = {'historial_academico'      : historial_academico,
                 'grado'                    : grado,
                 'repitencias_y_deserciones': repitencias_y_deserciones         }
    return historial
```

4. Clase Persona con toda la info

```python
class Persona:
    def __init__(self, doc_identidad = 12345678):
        # cada Persona tiene un doc_identidad único que lo diferencia de las demás instancias
        # Aunque se le puede dar uno provisional (12345678) para demostraciones
        info = generar_info()
        indice_desercion_dep = info['indice_desercion_dep']
        historial = generar_historial_academico(indice_desercion_dep, info['edad'])

        self.doc_identidad              = doc_identidad
        self.nombre                     = info['nombre']
        self.edad                       = info['edad']
        self.sexo                       = info['sexo']
        self.departamento_code          = info['departamento_code']
        self.departamento               = info['departamento']
        self.municipio                  = info['municipio']
        self.historial_academico        = historial['historial_academico']
        self.grado                      = historial['grado']
        self.repitencias_y_deserciones  = historial['repitencias_y_deserciones']
        
    #==
    def __eq__ (self, __o: object) -> bool:
        return (self.edad == __o.edad) #and (self.sexo == __o.sexo)
    #<
    def __lt__ (self, __o: object) -> bool:
        return (self.edad < __o.edad) #and (self.sexo == __o.sexo)
    #<=
    def __le__ (self, __o: object) -> bool:
        return (self.edad <= __o.edad) #and (self.sexo == __o.sexo)
    #>
    def __gt__ (self, __o: object) -> bool:
        return (self.edad > __o.edad) #and (self.sexo == __o.sexo)
    #>=
    def __ge__ (self, __o: object) -> bool:
        return (self.edad >= __o.edad) #and (self.sexo == __o.sexo)
    #!=
    def __ne__ (self, __o: object) -> bool:
        return (self.edad != __o.edad) #or (self.sexo != __o.sexo)
    #str format
    def __str__(self):
        return f'Objeto Persona=[doc identidad: {self.doc_identidad}, nombre: {self.nombre}, edad: {self.edad}, sexo: {self.sexo}, código dpto: {self.departamento_code}, dpto: {self.departamento}, municipio: {self.municipio}, grado: {self.grado}, repitencias y deserciones: {self.repitencias_y_deserciones}]'
```

5. Generador de Personas (las retorna en array de numpy)

```python
# El número mínimo de personas especificado en el ejercicio son 5 millones        
def generar_personas(numero_deseado) -> np.ndarray :
    lista_de_personas = deque()
    # Al agregarle 1000000 desde el principio, todos tienen el mismo número de
    # cifras. Además se parece más a una identificación real.h
    for persona_n in range(1000000,1000000+numero_deseado):
        doc_identidad = persona_n + 10000000  
        lista_de_personas.append(Persona(doc_identidad))

    return np.array(lista_de_personas)
```

6. Filtrador para solicitar sólo la información a usar de esas personas. Puede filtrar dependiendo del departamento. 

  Retorna un array de numpy hecho de tuplas con las 3 variables int que se ingresan en la red neuronal: ->   [ (int, int, int), (int, int, int), ...]

```python
# Para cada persona, le extrae su edad, grado y # de repitencias si pertenece a cierto departamento.
# Si no se ingresa departamento, se le hará la extracción a todas las personas sin excepción.
def filtracion(lista , departamento : int = None) -> np.ndarray :
    lista_filtrada = deque()

    for persona in lista:
        if persona.departamento_code == departamento or departamento == None:
            lista_filtrada.append([persona.edad,persona.grado,persona.repitencias_y_deserciones])

    return np.array(lista_filtrada)
```

7. Definimos y compilamos la red neuronal. 

  3 nodos de input, una sola capa intermedia de 6 nodos, y un solo nodo de output en este caso.

```python
n_entrada = 3
n_nodos = 6
n_salida = 1

model = Sequential()
model.add(Dense(n_nodos,  input_dim  = n_entrada,  activation = 'relu'))
model.add(Dense(n_salida))  
# Sigmoid sólo sirve para ejercicios binarios, no es adecuado para este caso.

model.compile(loss='mean_squared_error', optimizer='adam', metrics=[tf.keras.metrics.MeanSquaredLogarithmicError(name='Error Cuadrado Logarítmico (MSLE)')])
```

9. Categorizamos en rangos los outputs que nos dé la predicción, y luego retornamos el porcentaje de personas que están en cada uno

```python
def categorizador(prediccion):
    contador_bajo, contador_medio, contador_alto = 0 , 0 , 0
    total = len(prediccion)
    for i in prediccion:
        if 0 <= i[0] and i[0] < 0.333:
            contador_alto += 1
        elif 0.333 <= i[0] and i[0] < 0.666:
            contador_medio += 1
        else:
            contador_bajo += 1
    porcentaje_bajo   = contador_bajo * 100 / total
    porcentaje_medio  = contador_medio * 100 / total
    porcentaje_alto   = contador_alto * 100 / total

    return porcentaje_bajo , porcentaje_medio , porcentaje_alto
```

11. Graficación

  Puede graficar 3 barras de riezgo (bajo, medio, alto) si se analiza toda Colombia, o un departamento.

  Puede graficar una sola barra de riezgo si se analiza a una sola persona.

```python
def graficador(resultados):

    fig, ax = plt.subplots()
    ax.set_ylim([0, 100])
    ax.grid(axis = 'y', color = 'gray', linestyle = 'dashed')

    if len(resultados) == 1:
        ax.bar(['',' ','  '],[0,resultados[0],0], color='blue')
        plt.xlabel('Riezgo de deserción' , fontdict = {'fontsize':14, 'fontweight':'bold', 'color':'tab:blue'})
        plt.ylabel('Porcentaje (%)' , fontdict = {'fontsize':14, 'fontweight':'bold', 'color':'tab:blue'})

        

    if len(resultados) == 3:
        ax.bar(['BAJO','MEDIO','ALTO'],[resultados[0],0,0], color='green')
        ax.bar(['BAJO','MEDIO','ALTO'],[0,resultados[1],0], color='yellow')
        ax.bar(['BAJO','MEDIO','ALTO'],[0,0,resultados[2]], color='red')
        plt.xlabel('Categoría de riezgo' , fontdict = {'fontsize':14, 'fontweight':'bold', 'color':'tab:blue'})
        plt.ylabel('Porcentaje de población (%)' , fontdict = {'fontsize':14, 'fontweight':'bold', 'color':'tab:blue'})

    plt.show()
```
